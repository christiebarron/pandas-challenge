# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Starter_Code/Resources/schools_complete.csv"
student_data_to_load = "Starter_Code/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete["school_name"].unique())
print(f"There are {school_count} unique schools within this district.")
```

```python
# Calculate the total number of students
student_count = school_data_complete.nunique().loc["Student ID"]
print(f"There are {student_count} students in the district.")

#note get different results for unique names of students.
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
print(f"The district's total budget is ${total_budget:,d}")

school_data
#school_data_complete
```

```python
# Calculate the average (mean) math score
average_math_score_stu = school_data_complete["math_score"].mean()
average_math_score_sch = round(school_data_complete.groupby(["school_name"])[["math_score"]].mean(), 2)

print(f"The average math score grouped by student is: {round(average_math_score_stu,2)}")
print(f"The average math scores grouped by school are:")
print(average_math_score_sch)
```

```python
# Calculate the average (mean) reading score
average_read_score_stu = school_data_complete["reading_score"].mean()
average_read_score_sch = round(school_data_complete.groupby(["school_name"])[["reading_score"]].mean(), 2)

print(f"The average reading score grouped by student is: {round(average_read_score_stu,2)}")
print(f"The average reading scores grouped by school are:")
print(average_read_score_sch)
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
print(f"Approximately {round(passing_math_percentage,2)}% of students passed math")
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
passing_read_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_read_percentage = passing_read_count / float(student_count) * 100
print(f"Approximately {round(passing_read_percentage,2)}% of students passed reading")
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
print(f"Approximately {round(overall_passing_rate, 2)}% of students passed both reading and math")
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame([school_count, student_count, total_budget, average_math_score_stu, average_read_score_stu, passing_math_count, passing_math_percentage, passing_read_count, passing_read_percentage, overall_passing_rate]) #"value")
district_summary

district_summary = {
    "summary_value" : [school_count, student_count, total_budget, average_math_score_stu, average_read_score_stu, passing_math_count, passing_math_percentage, passing_read_count, passing_read_percentage, overall_passing_rate]
}
district_summary
# Formatting
#district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
#district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
#district_summary
```

```python
# Calculate the average test scores
per_school_math = school_data_complete.groupby(["school_name"])["math_score"].mean()
per_school_reading = school_data_complete.groupby(["school_name"])["reading_score"].mean()
#per_school_math.rename(columns = {"math_score": "math_average"})


school_data = pd.merge(school_data, per_school_math, on = "school_name", how = "left")
school_data = school_data.rename(columns = {"math_score": "average_math_score"})
school_data = pd.merge(school_data, per_school_reading, on = "school_name", how = "left")
school_data = school_data.rename(columns = {"reading_score" : "average_reading_score"})
```

```python
# Calculate the number of schools with math scores of 70 or higher
bins = [0, 70, 130]
labels = [0, 1]
school_passing_math = pd.cut(per_school_math, bins = bins, labels = labels)
print(f"The number of schools with math scores of 70 or higher is: {sum(school_passing_math)}")

school_data = pd.merge(school_data, school_passing_math, on = "school_name", how = "left")
school_data = school_data.rename(columns = {"math_score": "average_math_pass"})
school_data
```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = pd.cut(per_school_reading, bins = bins, labels = labels)
print(f"The number of schools with reading scores of 70 or higher is: {sum(school_passing_reading)}")

school_data = pd.merge(school_data, school_passing_reading, on = "school_name", how = "left")
school_data = school_data.rename(columns = {"reading_score": "average_read_pass"})
school_data
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
school_data["math_and_read_pass"] = 0
school_data.loc[(school_data["average_math_score"] >= 70) & (school_data["average_reading_score"]>=70), "math_and_read_pass"] = 1 
school_data

#passing_math_and_reading = school_data_complete[
#    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
#]
#passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
bins = [0, 70, 130]
labels = [0, 1]
school_data_complete["reading_pass"] = pd.cut(school_data_complete["reading_score"], bins = bins, labels = labels).astype(int)
school_data_complete["math_pass"] = pd.cut(school_data_complete["math_score"], bins = bins, labels = labels).astype(int)
school_data_complete["read_and_math_pass"] = 0
school_data_complete.loc[(school_data_complete["reading_pass"] == 1) & (school_data_complete["math_pass"] == 1) , "read_and_math_pass"] = 1
school_data_complete

df = school_data_complete.groupby("school_name")[["reading_pass", "math_pass","read_and_math_pass"]].sum().astype(float)
df = df.rename(columns= {"reading_pass": "sum_reading_pass", "math_pass": "sum_math_pass", "read_and_math_pass": "sum_read_and_math_pass"})
school_data = pd.merge(school_data, df, on = "school_name", how = "left")

df = school_data_complete.groupby("school_name")[["reading_pass", "math_pass","read_and_math_pass"]].mean().astype(float)
df = df.rename(columns= {"reading_pass": "mean_reading_pass", "math_pass": "mean_math_pass", "read_and_math_pass": "mean_read_and_math_pass"})
school_data = pd.merge(school_data, df, on = "school_name", how = "left")

school_data
#school_data_complete.groupby(["school_name"]).count()["student_name"] / per_school_counts
#school_data

#per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
#per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
#overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
```
