# PyCity Schools Analysis

## Executive Summary

This report articulates the characteristics and academic performance of the school district for the current academic year. It starts by providing high-level district summary information. It then provides summary information at the school level. It concludes with descriptive statistics describing school performance based on school budget, school size, and school type.

Primary takaways from this report is that the schools in this district have notable variability in their academic performance. Some of this variability seems to be tied to school type, with district schools having notably lower performance in math. Large schools also tend to have moderately lower performance in math. Inferential statistics are required for further insight.

#### District Summary

* The school district has a total of 15 unique schools, 39170 students, and a total budget of $24,649,428. 
* This year, approximately 74.95% students passed math, 85.81% of students passed reading, and 65.17% of students passed both math and reading.

#### School Summary

* The average school budget per student ranged from $578 to $655.
* All schools demonstrated average math and reading scores that exceeded the 70% cut-off.
* Unsurprisingly, the highest performing schools were all charter schools. These schools all had greater than 90% of students passing reading, greater than 90% students passing math, and approximately 84% of students passing reading and math.
* The lowest performing schools were all district schools. These schools had approximately 78% of students passing reading, approximately 64% of students passing math, and approximately 50% of students passing both reading and math.
* The average percentage of students passing reading and math tended to remain consistent across grades. 

#### Predictors of School Performance

* **Budget**: Counterintuitively, schools with higher budgets, did not yield better test results. On the contrary, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student). This is likely due to confounding variables such as the socioeconomic status of schools. For instance, schools with large numbers of struggling students due to poor socioeconomic conditions may recieve greater funding to better support their students. 

* **School Size**: On the whole, smaller size and medium-sized schools performed similarly to each other. However, both smaller and medium-size dschools dramatically out-performed large sized schools on percent passing math (90-91% versus 68%), and percent passing reading and math passing math performances (84% passing vs 55%). 

* **School Type**: In general, charter schools out-performed the public district schools across all metrics. This effect was more pronounced in math and overall passing rates than reading. Due to several confounding variables, the cause for this effect is uncertain. Charter schools likely have higher SES students, which is one explanation. Additional analysis using methods such as propensity score matching will be required to make valid causal claims, though even multiple regression controlling for school size and school budget would help elucidate this relationship.

# Statistical Analysis

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Starter_Code/PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "Starter_Code/PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete["school_name"].unique())
print(f"There are {school_count} unique schools within this district.")
```

```python
# Calculate the total number of students
student_count = school_data_complete.nunique().loc["Student ID"]
print(f"There are {student_count} students in the district.")

#note get different results for unique names of students.
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
print(f"The district's total budget is ${total_budget:,d}")

#print the full school data file
school_data
```

```python
# Calculate the average (mean) math score
average_math_score_stu = school_data_complete["math_score"].mean()
average_math_score_sch = round(school_data_complete.groupby(["school_name"])[["math_score"]].mean(), 2)

print(f"The average math score grouped by student is: {round(average_math_score_stu,2)}")
print(f"The average math scores grouped by school are:")
print(average_math_score_sch)
```

```python
# Calculate the average (mean) reading score
average_read_score_stu = school_data_complete["reading_score"].mean()
average_read_score_sch = round(school_data_complete.groupby(["school_name"])[["reading_score"]].mean(), 2)

print(f"The average reading score grouped by student is: {round(average_read_score_stu,2)}")
print(f"The average reading scores grouped by school are:")
print(average_read_score_sch)
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
print(f"Approximately {round(passing_math_percentage,2)}% of students passed math")
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
passing_read_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_read_percentage = passing_read_count / float(student_count) * 100
print(f"Approximately {round(passing_read_percentage,2)}% of students passed reading")
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
print(f"Approximately {round(overall_passing_rate, 2)}% of students passed both reading and math")
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame

#create dictionary
data = {"Total Schools" : school_count,
        "Total Students" : student_count,
        "Total Budget" : total_budget,
       "Average Math Score" : average_math_score_stu,
        "Average Reading Score": average_read_score_stu,
        "Passing Math Count": passing_math_count,
        "% Passing Math" : passing_math_percentage,
        "Passing Reading Count" : passing_read_count,
        "% Passing Reading" : passing_read_percentage,
        "% Overall Passing": overall_passing_rate,
    }

#convert to dataframe
district_summary = pd.DataFrame(data, index = [0]) 


# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary

```

```python
# Calculating passing rates
bins = [0, 70, 130]
labels = [0, 1]
school_data_complete["reading_pass"] = pd.cut(school_data_complete["reading_score"], bins = bins, labels = labels).astype(int)
school_data_complete["math_pass"] = pd.cut(school_data_complete["math_score"], bins = bins, labels = labels).astype(int)
school_data_complete["read_and_math_pass"] = 0
school_data_complete.loc[(school_data_complete["reading_pass"] == 1) & (school_data_complete["math_pass"] == 1) , "read_and_math_pass"] = 1
school_data_complete

df = school_data_complete.groupby("school_name")[["reading_pass", "math_pass","read_and_math_pass"]].sum().astype(float)
df = df.rename(columns= {"reading_pass": "sum_reading_pass", "math_pass": "sum_math_pass", "read_and_math_pass": "sum_read_and_math_pass"})
school_data = pd.merge(school_data, df, on = "school_name", how = "left")

df = school_data_complete.groupby("school_name")[["reading_pass", "math_pass","read_and_math_pass"]].mean().astype(float)
df = df.rename(columns= {"reading_pass": "mean_reading_pass", "math_pass": "mean_math_pass", "read_and_math_pass": "mean_read_and_math_pass"})
school_data = pd.merge(school_data, df, on = "school_name", how = "left")

school_data
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df["budget_capita"] = school_spending_df["budget_capita"].replace('[\$,]', '', regex=True).astype(float)
school_spending_df["% Passing Reading"] = school_spending_df["% Passing Reading"].replace('[\%,]', '', regex=True).astype(float)
school_spending_df["% Passing Math"] = school_spending_df["% Passing Math"].replace('[\%,]', '', regex=True).astype(float)
school_spending_df["% Overall Passing"] = school_spending_df["% Overall Passing"].replace('[\%,]', '', regex=True).astype(float)
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["budget_capita"], labels = labels, bins = spending_bins)
school_spending_df

school_spending_df = school_spending_df.loc[:,['School ID', 'school_name', 'type', 'size', 'budget', 'budget_capita','average_math_score', 'average_reading_score', '% Passing Reading', '% Passing Math', '% Overall Passing', 'Spending Ranges (Per Student)']]

#check df
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["average_math_score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["average_reading_score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

```

```python
# Assemble into DataFrame
spending_summary = pd.concat([spending_math_scores, spending_reading_scores, spending_passing_math, spending_passing_reading, overall_passing_spending], axis = 1)

# Display results
spending_summary.columns = ['Average Math Score', 'Average Reading Score', '% Passing Math',
       '% Passing Reading', '% Overall Passing']
spending_summary
```
